```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

G = nx.Graph()
edges = nx.read_edgelist("C:\\Users\\Abdul\\Desktop\\Discrete Project\\Top10PL.txt")
print(nx.info(edges))




d = [] # create a set
for n in edges.nodes():
    d.append(edges.degree(n))
    
ec = [] # create a set
for e in edges.edges():
    if(edges.degree(e[0]) > edges.degree(e[1])):
        ec.append(edges.degree(e[0]))
    else:
        ec.append(edges.degree(e[1]))
    
pos = nx.spring_layout(edges,k=0.3,iterations=80)

factor = 120 # to change the size of nodes with respect to their degree


# choose color plate from https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
# example: viridis
nx.draw(edges, pos,with_labels=True,font_color='#bbffee',
        edge_color = ec,  edge_cmap=plt.cm.plasma,  # edge color
        node_color = d, cmap=plt.cm.viridis, # node color
        node_size=[x * factor for x in d]) # node sizse
fig = plt.gcf()
fig.set_size_inches((10,14))





fig = plt.gcf()

nx.draw(G)
fig.suptitle('Top 10 Programming Languages',size=20,color="white")
fig.patch.set_facecolor("black")


plt.show()
```

```python
#degeree plotting
def plot_degree_distribution(wiki):
    degs = {}
    for n in wiki.nodes():
        deg = wiki.degree(n)

        if deg not in degs:
            degs[deg] = 0

        degs[deg] += 1
    items = sorted(degs.items())

    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot([k for (k, v) in items], [v for (k, v) in items])
    ax.set_xscale('log')
    ax.set_yscale('log')
    plt.title("Degree Distribution")
#     fig.savefig("C:\\Users\\Abdul\\Desktop\\degree_distribution.png")

plot_degree_distribution(edges)
```
