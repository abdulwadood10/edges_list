```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

G = nx.Graph()
edges = nx.read_edgelist("C:\\Users\\Abdul\\Desktop\\abcd.txt")
print(nx.info(edges))
print("Diameter:",nx.diameter(edges))
print("Clustering",nx.average_clustering(edges))
pos = nx.spring_layout(edges,k=0.5,iterations=200)
print(nx.average_shortest_path_length(edges))
nx.draw_networkx(edges,pos=pos,edge_color='c',node_color='brown',node_size=80,font_color="white",  font_size=10)
fig = plt.gcf()
fig.set_size_inches((15,15))


fig = plt.gcf()

nx.draw(G)
fig.patch.set_facecolor('black')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

G = nx.Graph()
edges = nx.read_edgelist("C:\\Users\\Abdul\\Desktop\\8.txt")
edges1 = nx.read_edgelist("C:\\Users\\Abdul\\Desktop\\7.txt")
edges2 = nx.read_edgelist("C:\\Users\\Abdul\\Desktop\\6.txt")
edges3 = nx.read_edgelist("C:\\Users\\Abdul\\Desktop\\5.txt")
edges4 = nx.read_edgelist("C:\\Users\\Abdul\\Desktop\\4.txt")

print(nx.info(edges))
print(nx.info(edges1))
print(nx.info(edges2))
print(nx.info(edges3))
print(nx.info(edges4))

pos = nx.spring_layout(edges,k=0.1,iterations=90)
pos1 = nx.spring_layout(edges1,k=0.1,iterations=50)
pos2 = nx.spring_layout(edges2,k=0.1,iterations=60)
pos3 = nx.spring_layout(edges3,k=0.1,iterations=70)
pos4 = nx.spring_layout(edges4,k=0.1,iterations=100)

nx.draw_networkx(edges,pos=pos,loc='middle',node_color="White",width=2.0,node_size=30,edge_color='blue', style='dashed', alpha=1.0,font_color="White",font_size=18)
nx.draw_networkx(edges1,pos1=pos1,node_color="White",width=2.0,node_size=30,edge_color='white', style='dotted', alpha=1.0,font_color="White",font_size=18)
nx.draw_networkx(edges2,pos2=pos2,node_color="White",width=2.0,node_size=30,edge_color='red', style='solid', alpha=1.0,font_color="White",font_size=18)
nx.draw_networkx(edges3,pos3=pos3,node_color="White",width=2.0,node_size=30,edge_color='green', style='dashdot', alpha=1.0,font_color="White",font_size=18)
nx.draw_networkx(edges4,pos3=pos3,node_color="White",width=2.0,node_size=30,edge_color='orange', style='dotted', alpha=1.0,font_color="White",font_size=18)
fig = plt.gcf()
fig.set_size_inches((20,18))
fig.suptitle('THE NETWORK GRAPH',size=36,color="white")


fig = plt.gcf()

nx.draw(G)
fig.patch.set_facecolor('xkcd:black')


plt.show()
```
